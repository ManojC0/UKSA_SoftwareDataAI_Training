### Some Dataframe practice

```python
# This is just some code to make a DF which we will use for practice

def func_makedata(a):
    ''' this is just a function to make some data
    for our spreasheet (or 'dataframe') - you can happily ignore this
    '''
    x1 = a**2
    y1 = np.cos(a)
    z1 = 3*a**2
    return x1, y1, z1

aa = np.linspace(0.,10.,5)
x1, y1, z1 = func_makedata(aa)

data_dict = {}
data_dict['aa'] = aa
data_dict['x1'] = x1
data_dict['y1'] = y1
data_dict['z1'] = z1

# Here I am making a dataframe from arays which have been put into a dictionary
# you don't need to know at this stage what a 'dictionary' is.
df = pd.DataFrame(data_dict)
df
```

Print column 'aa'

```python
df['aa']
```

```python
df['aa'][0]
```
